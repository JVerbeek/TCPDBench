```python
import os
import glob
import json
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import subprocess as subp
from statsmodels.tsa.stattools import adfuller, zivot_andrews
```

```python
for i, dataset in enumerate(glob.glob("*.json")):
    with open(dataset) as f:
        ds = json.load(f)
        try:
            x, y = np.array(ds["time"]["index"]), np.array(ds["series"][0]["raw"])
        except KeyError:
            print(ds.keys())
        ts = pd.Series(y).cumsum()
        print(pd.Series(y).describe())
        ma = ts.rolling(window=int(len(x)/5)).mean()
        ms = ts.rolling(window=int(len(x)/5)).var()
        plt.plot(x, y)
        plt.title(dataset)
        plt.show()
        fig, ax = plt.subplots(1, 3)
        ax[0].plot(ma.values)
        ax[0].set_title("moving_average")
        ax[1].plot(ms.values)
        ax[1].set_title("moving_variance")
        ax[2].hist(y[y != np.array(None)], bins=50)
        ax[2].set_title("histogram")
        adf = adfuller(y[y != np.array(None)])
        res =  adf[1] <= 0.05 
        print(f"Augmented Dickey-Fuller statistic for unit root: {adf[0]} \n with p-value {adf[1]}, \n stationary: {res}." )
        print("-"*50)
        plt.show()
```
