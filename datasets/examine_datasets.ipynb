```python
import os
import glob
import json
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import subprocess as subp
from statsmodels.tsa.stattools import adfuller, zivot_andrews
```

Most changes are abrupt changes as per the paper. In table C1 of the paper there is a table with min, max and avg CPs, as well as length and sampling frequency.

Thus, we miss some data with slower or more gradual changes.

We visualize all datasets and visualize the rolling mean, std, and some descriptive statistics (pandas' describe). We test stationarity by testing for a unit root with the Zivot Andrews test, which is suitable for testing that in the presence of one structural break, and the augmented Dickey-Fuller test, which tests the same thing 

```python
with open("/home/janneke/TCPDBench/analysis/annotations/annotations.json") as f:
    annot = json.load(f)

for i, dataset in enumerate(glob.glob("*.json")):
    with open(dataset) as f:
        ds = json.load(f)
    try:
        an = annot[dataset[:-5]]
        print("Average number of annotated CPs:", np.array([len(v) for v in an.values()]).mean())
    except KeyError:
        print(dataset[:-5])
    try:
        x, y = np.array(ds["time"]["index"]), np.array(ds["series"][0]["raw"])
    except KeyError:
        print(ds.keys())
    ts = pd.Series(y).cumsum()
    print(pd.Series(y).describe())
    ma = ts.rolling(window=int(len(x)/5)).mean()
    ms = ts.rolling(window=int(len(x)/5)).std()
    plt.plot(x, y)
    plt.title(dataset)
    plt.show()
    fig, ax = plt.subplots(1, 3)
    ax[0].plot(ma.values)
    ax[0].set_title("moving_average")
    ax[1].plot(ms.values)
    ax[1].set_title("moving_variance")
    ax[2].hist(y[y != np.array(None)], bins=50)
    ax[2].set_title("histogram")
    adf = adfuller(y[y != np.array(None)])
    za = zivot_andrews(y[y!= np.array(None)])
    res =  adf[1] <= 0.05 
    za_res = za[1] <= 0.05
    print(f"Zivot-Andrews test for unit root: {za[0]} \n with p-value {za[1]}, stationary: {za_res}")
    print(f"Augmented Dickey-Fuller statistic for unit root: {adf[0]} \n with p-value {adf[1]}, \n stationary: {res}." )
    print("-"*50)
    plt.show()
```
